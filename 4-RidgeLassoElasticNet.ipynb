```python
##Train test split
```

```python
df2.head()
```

```python
y = df2[["Classes_fire"]]
X = df2.drop("Classes_fire",axis=1)
```

```python
X
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2,random_state=15)
```

```python
#Scaler
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
Lregression = LinearRegression()
```

```python
Lregression.fit(X_train,y_train)
```

```python
##prediction
```

```python
y_pred = Lregression.predict(X_test)
```

```python
print(y_pred)
```

```python
score = r2_score(y_test,y_pred)
print(score)
```

```python
from sklearn.linear_model import ElasticNetCV
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
elastic = ElasticNetCV(
    l1_ratio=[.1, .5, .7, .9, .95, .99, 1],   # L1 (Lasso) + L2 (Ridge) karışımı
    alphas=np.logspace(-4, 4, 50),            # alpha değer aralığı
    cv=5,                                     # 5-fold cross validation
    random_state=42
)
```

```python
elastic.fit(X_train,y_train)
```

```python
y_pred_elastic = elastic.predict(X_test)
```

```python
print("R² Skor:", r2_score(y_test, y_pred_elastic))
print("MSE:", mean_squared_error(y_test, y_pred_elastic))
print("MAE:", mean_absolute_error(y_test, y_pred_elastic))
```
